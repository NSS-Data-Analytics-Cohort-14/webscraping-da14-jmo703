```python
import requests
```

```python
import pandas as pd
```

```python
URL = 'https://realpython.github.io/fake-jobs/'
response = requests.get(URL)
```

```python
response.status_code
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text)
```

```python
print(soup.prettify())
```

```python
type(soup.findall('h2 class'))
```

```python
soup.find('title is-5').text
```

```python
jobs = soup.findAll ("h2", class_='title is-5')
```

```python
for title in jobs:
    print(title.get_text(strip=True))
```

```python
job_titles = [tag.get_text(strip=True) for tag in jobs]
```

```python
print(job_titles)
```

```python
companies = soup.findAll ("h3", class_='subtitle is-6 company')
```

```python
for company in companies:
    print(company.get_text(strip=True))
```

```python
company_titles = [tag.get_text(strip=True) for tag in companies]
```

```python
print(company_titles)
```

```python
locations = soup.findAll ("p", class_='location')
```

```python
for location in locations:
    print(location.get_text(strip=True))
```

```python
location_titles = [tag.get_text(strip=True) for tag in locations]
```

```python
print(location_titles)
```
